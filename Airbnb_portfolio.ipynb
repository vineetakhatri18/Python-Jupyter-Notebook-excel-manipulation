# AirBnb Listing Analysis

### Objective 1: Data Prep & QA

```python
import pandas as pd

listings = pd.read_csv(
    "C:\Vineeta\_python practise\Airbnb+Data\Listings.csv", 
    encoding="ISO-8859-1", 
    low_memory=False,
    parse_dates=['host_since']

)
```

```python
listings.head()

```

```python
listings.info()
```

```python
paris_listings = (
    listings
    .query("city == 'Paris'")
    .loc[:, ["host_since", "neighbourhood", "city", "accommodates", "price"]]
)

paris_listings.info()
```

```python
paris_listings.isna().sum()
```

```python
paris_listings.describe()
```

```python
paris_listings.query("accommodates == 0 and price == 0").count()
```

# Objective 2: Preparation for visualization



```python
price_listings_neighbourhood = (
    paris_listings
    .groupby("neighbourhood")
    .agg({"price" : "mean"})
    .sort_values("price")
)

price_listings_neighbourhood.tail()
```

```python
price_listings_accommodates = (
    paris_listings
    .query("neighbourhood == 'Elysee'")
    .groupby("accommodates")
    .agg({"price" : "mean"})
    .sort_values("price")
)

price_listings_accommodates.tail()
```

```python
price_listings_overtime = (
    paris_listings
    .set_index("host_since")
    .resample("Y")
    .agg({
        "neighbourhood": "count",
        "price" : "mean"})
)

price_listings_overtime.head()
```

# Objective 3: Visualizing data

```python
import seaborn as sns

(price_listings_neighbourhood
.plot
.barh(
    title = "Average Listing Price by Paris Neighbourhood",
    xlabel = "Price per Night (Euros)",
    ylabel = "Neighbourhood",
    legend = None
)
)

sns.despine()
```

```python
(price_listings_accommodates
.plot
.barh(
    title = "Average Listing Price by Accommodation Number",
    xlabel = "Price per Night (Euros)",
    ylabel = "Accommodation Capecity",
    legend = None
)
)

sns.despine()
```

```python
price_listings_overtime["neighbourhood"].plot(
    ylabel= "New Hosts",
    title = "New AirBnb Hosts in Paris Over Time"
)

sns.despine()
```

```python
price_listings_overtime["price"].plot(
    ylabel= "Average Price (Euros)",
    title = "Average AirBnb Price in Paris Over Time"
)

sns.despine()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.plot(
    price_listings_overtime.index,
    price_listings_overtime["neighbourhood"],
    label= "New Hosts",
    c= "pink"
)

ax.set_ylabel("New Hosts")

ax2 = ax.twinx()

ax2.plot(
    price_listings_overtime.index,
    price_listings_overtime["price"],
    label= "Average Price"
)
ax2.set_ylim(0)

ax2.set_ylabel("Average Price")

ax.set_title("2015 Regulations Lead to Fewer Hosts, Higher Prices")
```
